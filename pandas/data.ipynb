```python
import pandas as pd
path = '/Users/leeyu/Downloads/45071_1.csv'
columns = ['date', 'x', 'y', 'z']
```

```python
%%time
df = pd.read_csv(path, names=columns, header=None)
```

```python
df
```

```python
%%time
df2 = pd.DataFrame()
now = df.iloc[0, 0][:19]
x_sum = df.iloc[0, 1]
y_sum = df.iloc[0, 2]
z_sum = df.iloc[0, 3]
count = 1

for i in range(1, df.shape[0]):
    if now == df.iloc[i, 0][:19]:
        count += 1
        x_sum += df.iloc[i, 1]
        y_sum += df.iloc[i, 2]
        z_sum += df.iloc[i, 3]
    else:
        insert_list = [[now, x_sum / count, y_sum / count, z_sum / count]]
        df2 = df2.append(insert_list, ignore_index=True)
        now = df.iloc[i, 0][:19]
        x_sum = df.iloc[i, 1]
        y_sum = df.iloc[i, 2]
        z_sum = df.iloc[i, 3]
        count = 1
insert_list = [[now, x_sum / count, y_sum / count, z_sum / count]]
df2 = df2.append(insert_list, ignore_index=True)

df2.columns = ['date', 'x', 'y', 'z']
df2
```

```python
%%time
# df2 = df.iloc[0:100000, 0:4]
df2.to_csv("result.csv")
```
