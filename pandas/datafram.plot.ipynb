```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
dict1 = {'A': np.random.randint(53, 1000, 10),
        'B': np.random.randint(100, 392, 10),
        'C': np.random.randint(10, 120, 10)}
df = pd.DataFrame(dict1)
df.plot(kind='line', title='line'),
df.plot(kind='bar', title='bar'), 
df.plot(kind='barh', title='barh'), 
df.plot(kind='hist', title='hist'), 
df.plot(kind='box', title='box'),
df.plot(kind='kde', title='hist'),
df.plot(kind='density', title='density'), 
df.plot(kind='area', title='area'),
df.plot(kind='pie', title='pie', subplots=True),
df.plot(kind='scatter', title='scatter', x='A', y='B'),
df.plot(kind='hexbin', title='hexbin', x='A', y='B')

plt.show()
```
